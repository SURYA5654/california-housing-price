### Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
# Program for median imputation
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]
    return median
```

```python
# Calculate the median of median_income column
calculate_median(df['median_income'])
```
